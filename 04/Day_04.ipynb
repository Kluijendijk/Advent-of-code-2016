```julia
using DataStructures
p = r"([A-z-]*)(\d*)\[([A-z]*)\]"
inp = [match(p, x) for x in readlines("input.txt")]
inp = [[string(x[1]), parse(Int, x[2]), x[3]] for x in inp]
```

```julia
function countletters(str)
    d = DefaultDict(0)
    for ind in 1:length(str)
        l = str[ind]
        d[l] += 1
    end
    return sort(collect(d), by=x->x[2] - Int(x[1])/1000,rev=true) #subtracting int value of char/1000 to ensure alphabetical order
end


rooms = []

val = 0

for room in inp
    counts = countletters(replace(room[1], "-"=>""))
    first5 = join([x[1] for x in counts[1:5]])
    if first5 == room[3]
        val += room[2]
    end
end
val
```

```julia
function decrypt(room)
    s = room[1]
    ciph = room[2]%26
    out = ""
    for l in s
        if l == '-'
            out *= l
            continue
        end
        out *= Char((Int(l)-97+ciph)%26 + 97)
    end
    return out
end
for room in inp
    println("$(decrypt(room)): $(room[2])") #not told what the exact room name would be, :/
end
```
